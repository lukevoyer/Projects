## Setup

The following code imports the required libraries and loads a dataset containing information about diamonds into a Pandas DataFrame. Information about the dataset can be found [here](https://vincentarelbundock.github.io/Rdatasets/doc/ggplot2/diamonds.html).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
def normal_distribution(x, mu, sigma):
    return 1/(sigma * np.sqrt(2*np.pi)) * np.exp(-(x-mu)**2/(2*sigma**2))
```

```python
df = pd.read_csv('diamonds.csv')
df.pop('Unnamed: 0');
```

## Density of the Fair Diamond Prices

Make a plot that meets the following criteria:
* normalized histogram with 30 bins using the Fair cut diamond prices
* a line plot of the normal distribution using the mean and standard deviation of the Fair cut diamond prices
* appropriate labels on the both the x and y axes
* appropriate title
* appropriate legend

```python
# Initialize plot
fig,ax = plt.subplots(1,1)

# Locate subset of 'df' that specifies the prices of fair-cut diamonds
a = (df.loc[df['cut'] == 'Fair']['price']).to_numpy()

# Create normalized (density = True) histogram for the prices of fair-cut diamonds containing 30 bins
ax.hist(a, 30, density = True, label = "Frequency of Fair-Cut Diamonds Prices")

# Set labels + title
ax.set_xlabel('price of fair-cut diamonds in dollars')
ax.set_ylabel('normalized frequency')
ax.set_title('Frequency of Fair-Cut Diamond Prices vs. Fair-Cut Diamond Price Normal Distribution')

# Plot normal distribution of the prices of fair-cut diamonds
ax.plot(a,normal_distribution(a, np.mean(a), np.std(a)),label = "Fair-Cut Diamond Price Normal Distribution")

# Create legend
ax.legend(loc='upper center', bbox_to_anchor=(1.5, 0.8), shadow=True, ncol=1)

# Show plot
plt.show()
```

## Bar chart of average price per cut

Make a plot the meets the following criteria:
* horizontal bar chart of the mean prices of the diamond cuts
* ten evenly spaced tick marks on the x axis from 0 to the maximum mean price
* appropriate labels on the x and y axes
* appropriate title

```python
# Initialize plot
fig,ax = plt.subplots(1,1)

# Get individual cuts to create easier for loop statement
cuts = (df['cut'].drop_duplicates()).array

# Initialize mean prices list
mean_prices = []

# Loop over mean cut prices and store them in mean_prices
for x in cuts:
    mean_prices.append(np.mean((df.loc[df['cut'] == x]['price']).to_numpy()))
mean_prices = np.array(mean_prices)

# Create bar chart
ax.barh(cuts, mean_prices) 

# Set labels + title
ax.set_title('Mean Prices of Diamond Cuts')
ax.set_xlabel('mean price')
ax.set_ylabel('diamond cut')

# Define number of ticks for horizonatal bar chart
num_ticks = 10

# Set tick range
plt.xticks(np.arange(0, int(np.floor(np.amax(mean_prices))) , (round(int(np.floor(np.amax(mean_prices))), -3) / num_ticks)))

# Show plot
plt.show()
```
